```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
carpeta = "data"
ruta = os.path.join(carpeta, "dataset_tratado.csv")
df = pd.read_csv(ruta)
```

```python
df.columns
```

```python
df['job_level'].value_counts()
```

```python
# 1. Extraemos todo lo que esté entre comillas simples ' '
# El patrón r"'([^']+)'" captura el contenido sin las comillas
tecnologias = df['description'].str.extractall(r"'([^']+)'")

# 2. Contamos los valores
conteo_tech = tecnologias[0].value_counts()

print(conteo_tech)
```

```python
ofertasJunior = df[df['job_level'] == 'entry level']
#print(ofertasJunior)
tecnologias_junior = ofertasJunior['description'].str.extractall(r"'([^']+)'")

# 2. Contamos los valores
conteo_tech_junior = tecnologias_junior[0].value_counts()

print(conteo_tech_junior)
```

```python
plt.figure(figsize=(10, 6))
conteo_tech.head(10).plot(kind='bar')
plt.title('Top 10 Tecnologías más Requeridas')
plt.xlabel('Tecnología')
plt.ylabel('Número de Ofertas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
conteo_tech_junior.head(10).plot(kind='bar')
plt.title('Top 10 Tecnologías más Requeridas en juniors')
plt.xlabel('Tecnología')
plt.ylabel('Número de Ofertas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
